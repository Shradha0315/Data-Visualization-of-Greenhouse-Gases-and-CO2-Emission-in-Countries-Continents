```python
df1 = pd.read_csv('D:/DCU/CA682_DataVisualization/Assignment1/Dataset/Raw_DataSets/Countries and Continents.csv')
```

```python
df1.head()
```

```python
def unicode_string(var): 
    temp=''
    for i in var:
        if i != ' ':
            x=unidecode(i)
            temp+=x
        else:
            temp+=' '
    return temp
```

```python
country_updt=[]
for country in df1['Country']:
    country_clean=unicode_string(list(country))
    country_updt+=[country_clean]
```

```python
continent_updt=[]
for continent in df1['Continent']:
    continent_clean=unicode_string(list(continent))
    continent_updt+=[continent_clean]
```

```python
map_obj = dict(zip(country_updt,continent_updt))
```

```python
map_obj
```

```python
for i in range(len(df['country'])):
    if df['country'][i] in country_updt:
        df['continent'][i]=map_obj.get(df['country'][i])
```
